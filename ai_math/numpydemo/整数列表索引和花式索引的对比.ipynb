```python
import numpy as np
data = np.random.randint(10, 100, (5, 5))
```

```python
data
```

```python
list(range(4, -1, -1))
```

```python
list(range(5))
```

使用整数数组作为索引

```python
data[range(4, -1, -1), range(5)] = 0
```

```python
data
```

```python
data[range(5), range(5)] = 0
```

```python
data
```

使用花式索引对比整数索引

```python
data[np.ix_(range(5), range(5))] = 100
```

```python
data
```

```python
np.ix_(range(5), range(5))
```

用图片看的会更清楚一些

```python
import scipy.misc
import matplotlib.pyplot as plt
%pylab inline
img = scipy.misc.ascent()
```

```python
img.shape
```

```python
plt.imshow(img)
```

使用整数数组索引，加上一条黑色对角线

```python
rows = img.shape[0]
cols = img.shape[1]
img[range(rows), range(cols)] = 0
```

```python
plt.imshow(img)
```

```python
img[range(rows-1, -1, -1), range(cols)] = 0
```

```python
plt.imshow(img)
```

使用花式索引

```python
newimg2 = img.copy()
newimg2[np.ix_(range(rows), range(cols))] = 0
plt.imshow(newimg2)
```
