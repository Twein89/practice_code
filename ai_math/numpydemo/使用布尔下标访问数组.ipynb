```python
import numpy as np
import scipy.misc as misc
import matplotlib.pyplot as plt
%pylab inline
```

```python
img = misc.face()
img.shape
plt.imshow(img)
```

```python
rows = img.shape[0]
cols = img.shape[1]
```

```python
#

def get_indicies(size, skip_num=40):
    num = np.arange(size)
    return num % skip_num == 0
```

```python
new_row_idx = get_indicies(rows)
new_col_idx = get_indicies(cols)
```

```python
img_cp1 = img.copy()
img_cp1[new_row_idx, :] = [255, 0, 0]
img_cp1[:, new_col_idx] = [0, 255, 0]
plt.imshow(img_cp1)
```

```python
img_cp2 = img.copy()
new_row_idx = get_indicies(rows, 3)
new_col_idx = get_indicies(cols, 3)
img_cp2[np.ix_(new_row_idx, new_col_idx)] = [255, 0, 0]
plt.imshow(img_cp2)
```

```python
img_cp3 = img.copy()
img_cp3.max()
img_cp3[(img_cp3 > img_cp3.min()) & (img_cp3 < img_cp3.max() / 3)] = 80
img_cp3[(img_cp3 > img_cp3.max()/3) & (img_cp3 < 2 * img_cp3.max()/3)] = 160
img_cp3[(img_cp3 > 2 * img_cp3.max()/3) & (img_cp3 < 3 * img_cp3.max())] = 255
plt.imshow(img_cp3)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(img_cp1)
plt.subplot(223)
plt.imshow(img_cp2)
plt.subplot(224)
plt.imshow(img_cp3)
```
